Visualisation of researc

```python
import pandas as pd
import plotly.express as px
```

```python
# Reading in the CSV File with our Data
df = pd.read_csv("rq_3_richness_sh_data.csv")
```

We create a map which shows rural vs urban areas and the species richness of the location 


```python
# Creating an Map with urban and rural areas 
fig = px.scatter_map(
    df,
    lat="lat",
    lon="lon",
    # Areas with higher species richness are represented with larger circles 
    size="species_richness",
    color="area_type",
    color_discrete_map={
        "urban": "purple",
        "rural": "green"
    },
    zoom=5,
    #coordinates for center of Schleswig-Hosltein
    center={"lat": 54, "lon": 9.8},
)

fig.show()
```
